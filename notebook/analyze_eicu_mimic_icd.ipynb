```python
eicu_mimic_icd_data = eicu_mimic_patient_diagnosis[eicu_mimic_patient_diagnosis["hospitalid"].isin([2001, 1001])].iloc[:, 4:]
eicu_mimic_hospital_ids = eicu_mimic_patient_diagnosis[eicu_mimic_patient_diagnosis["hospitalid"].isin([2001, 1001])].iloc[:, 1]
eicu_icd_data = eicu_mimic_patient_diagnosis[eicu_mimic_patient_diagnosis["hospitalid"] == 1001].iloc[:, 4:]
mimic_icd_data = eicu_mimic_patient_diagnosis[eicu_mimic_patient_diagnosis["hospitalid"] == 2001].iloc[:, 4:]

eicu_mimic_hospital_ids_np = eicu_mimic_hospital_ids.to_numpy()
eicu_mimic_icd_data_np = eicu_mimic_icd_data.to_numpy()
eicu_icd_data_np = eicu_icd_data.to_numpy()
mimic_icd_data_np = mimic_icd_data.to_numpy()

icd_code_names = eicu_mimic_icd_data.columns
```

```python
import pandas as pd

def convert_icd9_to_disease(icd_9):
    if pd.isna(icd_9):
        return "Others"
    primary_icd9 = icd_9.split(',')[0].strip()
    try:
        # Convert the input to a float to handle both numeric and decimal ICD-9 codes
        icd_9_float = float(primary_icd9)

        # Check the ICD-9 code against the known ranges
        if 1 <= icd_9_float <= 139.9:
            return "Infection"
        elif 140 <= icd_9_float <= 239.9:
            return "Neoplasms"
        elif 240 <= icd_9_float <= 279.9:
            return "Endocrine"
        elif 280 <= icd_9_float <= 289.9:
            return "Blood"
        elif 290 <= icd_9_float <= 319:
            return "Mental"
        elif 320 <= icd_9_float <= 389.9:
            return "Nervous"
        elif 390 <= icd_9_float <= 459.9:
            return "Circulatory"
        elif 460 <= icd_9_float <= 519.9:
            return "Respiratory"
        elif 520 <= icd_9_float <= 579.9:
            return "Digestive"
        elif 580 <= icd_9_float <= 629.9:
            return "Genitourinary"
        elif 630 <= icd_9_float <= 676.9:
            return "Pregnancy"
        elif 680 <= icd_9_float <= 709.9:
            return "Skin"
        elif 710 <= icd_9_float <= 739.9:
            return "Musculoskeletal"
        elif 740 <= icd_9_float <= 759.9:
            return "Congenital"
        elif 760 <= icd_9_float <= 799.9:
            return "Perinatal"
        elif 800 <= icd_9_float <= 1000:
            return "Poisoning"
        elif icd_9.startswith("V"):
            return "Others"
        else:
            return "Others"
    
    except ValueError:
        return "Others"

disease_color_map = {
    "Infection": "#005896",
    "Neoplasms": "#dc5f00",      # SteelBlue
    "Endocrine": "#008002",      # LimeGreen
    "Blood": "#b40005",          # Crimson
    "Mental": "#74499c",         # DarkViolet
    "Nervous": "#6c382e",        # Gold
    "Circulatory": "#ab3db3",    # OrangeRed
    "Respiratory": "#2e2e2e",    # DarkTurquoise
    "Digestive": "#9c9c00",      # DeepPink
    "Genitourinary": "#009eac",  # MediumSlateBlue
    "Pregnancy": "#abcc25",      # HotPink
    "Skin": "#f06e60",           # SaddleBrown
    "Musculoskeletal": "#3bd156",# DarkOliveGreen
    "Congenital": "#c7b228",     # BlueViolet
    "Perinatal": "#ff5c7c",      # IndianRed
    "Poisoning": "#1268fd",      # DarkOrange
    "Others": "#696969",         # DimGray
    "Unknown": "#808080"         # Gray
}

hospital_color_map = {
    1001: "#1268fd",
    2001: "#ff5c7c",
}
```

```python
disease_labels = [convert_icd9_to_disease(x) for x in icd_code_names]
disease_labels
```

```python
import numpy as np

disease_labels = np.array(disease_labels)

active_disease_eicu = np.argmax(eicu_icd_data_np, axis=1)
no_active_disease_mask_eicu = np.all(eicu_icd_data_np == 0, axis=1)

patients_all_df_disease_eicu = disease_labels[active_disease_eicu]
patients_all_df_disease_eicu[no_active_disease_mask_eicu] = "Others"

active_disease_mimic = np.argmax(mimic_icd_data_np, axis=1)
no_active_disease_mask_mimic = np.all(mimic_icd_data_np == 0, axis=1)

patients_all_df_disease_mimic = disease_labels[active_disease_mimic]
patients_all_df_disease_mimic[no_active_disease_mask_mimic] = "Others"

active_disease_eicu_mimic = np.argmax(eicu_mimic_icd_data_np, axis=1)
no_active_disease_mask_eicu_mimic = np.all(eicu_mimic_icd_data_np == 0, axis=1)

patients_all_df_disease_eicu_mimic = disease_labels[active_disease_eicu_mimic]
patients_all_df_disease_eicu_mimic[no_active_disease_mask_eicu_mimic] = "Others"
```

```python
import umap.umap_ as umap
from matplotlib.lines import Line2D

from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score

import matplotlib.pyplot as plt

reducer = umap.UMAP(n_neighbors=10, n_components=2, min_dist=0.8,
                            random_state=np.random.RandomState(25),
                            transform_seed=np.random.RandomState(25))
principal_components = reducer.fit_transform(eicu_mimic_icd_data_np)
print(principal_components.shape)
```

```python
final_np = np.hstack((principal_components, patients_all_df_disease_eicu_mimic[:, None]))

legend_elements = []
for label in np.unique(patients_all_df_disease_eicu_mimic):
    scatter_rows = final_np[np.where(final_np[:, 2] == label)]
    plt.scatter(scatter_rows[:, 0].astype(float), scatter_rows[:, 1].astype(float), color=disease_color_map[label], s=0.02, alpha=0.2, label=label)
    legend_element = Line2D([0], [0], marker='o', color='w', markerfacecolor=disease_color_map[label], markersize=8, label=label)
    legend_elements.append(legend_element)

plt.title("K-Means Clustering of Diagnosis using ICD-9 Data (MIMIC-III & eICU)")
# plt.text(30, 17, f'ARI: {ari_score_diagnosis:.4f}', color='red', fontsize=12, bbox=dict(facecolor='yellow', alpha=1.0))
plt.legend(title="Disease", handles=legend_elements, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```

```python
final_np = np.hstack((principal_components, eicu_mimic_hospital_ids_np[:, None]))

legend_elements = []
for label in np.unique(eicu_mimic_hospital_ids_np):
    label_txt = "mimic" if label == 2001 else "eicu"
    scatter_rows = final_np[np.where(final_np[:, 2] == label)]
    plt.scatter(scatter_rows[:, 0].astype(float), scatter_rows[:, 1].astype(float), color=hospital_color_map[label], s=0.02, alpha=0.2, label=label_txt)
    legend_element = Line2D([0], [0], marker='o', color='w', markerfacecolor=hospital_color_map[label], markersize=8, label=label_txt)
    legend_elements.append(legend_element)

plt.title("K-Means Clustering of Diagnosis using ICD-9 Data (MIMIC-III & eICU)")
# plt.text(30, 17, f'ARI: {ari_score_diagnosis:.4f}', color='red', fontsize=12, bbox=dict(facecolor='yellow', alpha=1.0))
plt.legend(title="Hospitals", handles=legend_elements, loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
