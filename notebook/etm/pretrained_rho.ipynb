```python
import torch
import numpy as np
import random

# Set the seed for reproducibility
random_seed = 42
torch.manual_seed(random_seed)
np.random.seed(random_seed)
random.seed(random_seed)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

icd_data = pd.read_csv("../data/eicu_mimic_patient_diagnosis.csv")
eicu_hospital_data = icd_data[icd_data["hospitalid"] == 1001]
mimic_hospital_data = icd_data[icd_data["hospitalid"] == 2001]
eicu_train_data, _ = train_test_split(eicu_hospital_data, test_size=0.2, random_state=42)
mimic_train_data, _ = train_test_split(mimic_hospital_data, test_size=0.2, random_state=42)
train_data_total = pd.concat([eicu_train_data, mimic_train_data])
x_train = train_data_total.iloc[:, 4:].to_numpy()
x_bow_train = []
for row in x_train:
    word_id = list(np.where(row == 1)[0])
    x_bow_train.append(word_id)
x_bow_train
```

```python
from gensim.models import Word2Vec

# Train Word2Vec embeddings
model = Word2Vec(sentences=x_bow_train, vector_size=256, window=5, min_count=1, sg=1)
embedding = model.wv.vectors
```
